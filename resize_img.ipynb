```python
import cv2
import os
import glob
```

```python
test_images = glob.iglob("./data/test/" + '*.jpg', recursive=True)
train_images = glob.iglob("./data/train/" + '*/*/*.jpg', recursive=True)
test_images = list(test_images)
train_images = list(train_images)
print('test_images:', len(test_images))
print('train_images:', len(train_images))
```

```python
for i, f in enumerate(test_images):
    print(i, f)
    img = cv2.imread(f)
    newImg = cv2.resize(img, (112, 112))
    cv2.imwrite(f,newImg)

```

```python
for i, f in enumerate(train_images):
    print(i, f)
    img = cv2.imread(f)
    newImg = cv2.resize(img, (112, 112))
    cv2.imwrite(f,newImg)

```
